```python
import numpy as np
import pandas as pd
import itertools
import random
import time

start_time1 = time.time()
file_path = '/content/Adaptive_MLaaS_Composition_exp3.csv'
adaptive_mlaas_df = pd.read_csv(file_path)
adaptive_mlaas_df
adaptive_mlaas_df = adaptive_mlaas_df.rename(columns={'Service_ID': 'services'})

# Load the CSV

adaptive_mlaas_df=adaptive_mlaas_df[:2000]
# Simplified Parameters
group_size = 3  # Groups of 3 services for simplicity
alpha = 1  # Exploration-exploitation parameter
batch_size = 100  # Number of groups per batch
num_features = 4  # Based on the Feature_Vector size
max_combinations = 50000  # Limit to the first 2000 combinations

# Generate all possible groups
service_ids = adaptive_mlaas_df['services'].tolist()
all_possible_groups = list(itertools.combinations(service_ids, group_size))[:max_combinations]  # Limit to 2000 combinations
num_iterations = len(all_possible_groups) // batch_size  # Number of iterations based on batch size

# Initialize A and b for Each Group
A = {}
b = {}

# Function to Compute UCB using feature vectors from adaptive_mlaas_df
def compute_ucb(group, combined_feature_vector):
    if group not in A:
        A[group] = np.identity(num_features)
        b[group] = np.zeros(num_features)
    A_inv = np.linalg.inv(A[group])
    theta_hat = np.dot(A_inv, b[group])
    ucb_value = np.dot(combined_feature_vector, theta_hat) + alpha * np.sqrt(
        np.dot(combined_feature_vector, np.dot(A_inv, combined_feature_vector))
    )
    return ucb_value

# Learning Iteration
best_groups_with_ucb_and_rewards = []  # To store the best groups with their UCB and rewards

for iteration in range(num_iterations):
    start_index = iteration * batch_size
    end_index = start_index + batch_size
    sampled_groups = all_possible_groups[start_index:end_index]  # Sequentially pick 250 groups
    group_ucb_scores = {}
    for group in sampled_groups:
        feature_vectors = [
            np.fromstring(adaptive_mlaas_df.loc[adaptive_mlaas_df['services'] == s, 'Feature_Vector'].values[0].strip('[]'), sep=' ')
            for s in group
        ]
        combined_feature_vector = np.mean(feature_vectors, axis=0)
        group_ucb_scores[group] = compute_ucb(group, combined_feature_vector)

    # Select the best group based on UCB
    best_group = max(group_ucb_scores, key=group_ucb_scores.get)
    matched_scores = adaptive_mlaas_df[adaptive_mlaas_df['services'].isin(best_group)]['Composability_Score'].tolist()
    mean_score = np.mean(matched_scores)
    group_reward = mean_score if mean_score >= 0.45 else 0

    # Store the best group with its UCB and reward
    best_groups_with_ucb_and_rewards.append((best_group, group_ucb_scores[best_group], group_reward))

# Extract unique high-performing services from best groups
unique_high_ucb_services = set()
for group, ucb_score, reward in best_groups_with_ucb_and_rewards:
    unique_high_ucb_services.update(group)

# Further filter services with rewards greater than 0.8
high_reward_services = set()
for group, ucb_score, reward in best_groups_with_ucb_and_rewards:
    if reward > 0.55:
        high_reward_services.update(group)

# Generate new search space from unique high-performing services with rewards > 0.8
reduced_search_space = list(itertools.combinations(high_reward_services, group_size))
print(reduced_search_space)

# Perform combinatorial search on reduced search space
final_groups_with_ucb_and_rewards = []
for group in reduced_search_space:
    feature_vectors = [
        np.fromstring(adaptive_mlaas_df.loc[adaptive_mlaas_df['services'] == s, 'Feature_Vector'].values[0].strip('[]'), sep=' ')
        for s in group
    ]
    combined_feature_vector = np.mean(feature_vectors, axis=0)
    ucb_score = compute_ucb(group, combined_feature_vector)
    reward_scores = adaptive_mlaas_df[adaptive_mlaas_df['services'].isin(group)]['Composability_Score'].tolist()
    mean_reward = np.mean(reward_scores)
    final_groups_with_ucb_and_rewards.append((group, ucb_score, mean_reward))

# Sort the groups by reward in descending order
final_groups_with_ucb_and_rewards.sort(key=lambda x: x[2], reverse=True)

# Count reward categories
reward_categories = {
    "Excellent": 0,  # Reward >= 0.8
    "Good": 0,       # 0.6 <= Reward < 0.8
    "Average": 0,    # 0.45 <= Reward < 0.6
    "Poor": 0        # Reward < 0.45
}

for group, ucb_score, reward in final_groups_with_ucb_and_rewards:
    if reward >= 1:
        reward_categories["Excellent"] += 1
    elif 0.6 <= reward < 0.8:
        reward_categories["Good"] += 1
    elif 0.45 <= reward < 0.6:
        reward_categories["Average"] += 1
    else:
        reward_categories["Poor"] += 1

# Output all high-performing groups with their UCB scores, rewards, and count
print("Final High-Performing Groups with UCB and Rewards (Sorted Descending by Reward):")
for count, (group, ucb_score, reward) in enumerate(final_groups_with_ucb_and_rewards, 1):
    if reward >= 1:
        print(f"{count}. Group: {group}, UCB Score: {ucb_score}, Reward: {reward}")

# Output reward category counts
print("\nReward Category Counts:")
for category, count in reward_categories.items():
    print(f"{category}: {count}")
# End measuring time
end_time1 = time.time()

# Calculate computation time
computation_time1 = end_time1 - start_time1

print(f"\nTotal Computation Time: {computation_time1:.2f} seconds")
```

**Multi Arm Bandid**

```python
import time
import math
import random
from itertools import combinations
import pandas as pd

# Load dataset from CSV file
file_path = '/content/AdaptiveMLaaS.csv'
client_data = pd.read_csv(file_path)
client_data
# Assuming the dataset has columns: 'Client_ID' and 'composability score'
# Verify columns
if 'services' not in client_data.columns or 'Composability_Score' not in client_data.columns:
    raise ValueError("The dataset must contain 'services and 'composability score' columns.")

# Multi-Armed Bandit parameters
combination_size = 3
num_clients = len(client_data)
sampled_combinations = 10000  # Randomly sample 10,000 combinations
combinations_list = random.sample(list(combinations(client_data['services'], combination_size)), sampled_combinations)

# Initialize reward estimates and pull counts
reward_estimates = {comb: 0 for comb in combinations_list}
pull_counts = {comb: 0 for comb in combinations_list}
total_trials = 5000

# Function to compute mean reward (composability score) for a given combination
def calculate_mean_reward(comb):
    rewards = client_data[client_data['services'].isin(comb)]['Composability_Score'].values
    return rewards.mean()

# Start the MAB process
start_time = time.time()

for trial in range(1, total_trials + 1):
    ucb_values = {}
    for comb in combinations_list:
        if pull_counts[comb] == 0:
            ucb_values[comb] = float('inf')  # Force exploration
        else:
            average_reward = reward_estimates[comb]
            ucb_values[comb] = average_reward + math.sqrt(2 * math.log(trial) / pull_counts[comb])

    # Select the combination with the highest UCB value
    selected_combination = max(ucb_values, key=ucb_values.get)

    # Simulate getting a reward (use mean reward for the selected combination)
    observed_reward = calculate_mean_reward(selected_combination)

    # Update counts and reward estimates
    pull_counts[selected_combination] += 1
    n = pull_counts[selected_combination]
    reward_estimates[selected_combination] += (observed_reward - reward_estimates[selected_combination]) / n

# Filter and sort combinations with estimated rewards
sorted_combinations = sorted(reward_estimates.items(), key=lambda x: x[1], reverse=True)

# Output top 10 combinations
print("\nTop 10 Combinations with Highest Estimated Reward:")
for i, (combination, estimated_reward) in enumerate(sorted_combinations[:10], 1):
    print(f"{i}. Combination: {combination}, Estimated Reward: {estimated_reward:.2f}")

# Compute total execution time
end_time = time.time()
computation_time = end_time - start_time

print(f"\nTotal Computation Time: {computation_time:.2f} seconds")
```

```python
import random
from itertools import combinations
import time
import pandas as pd

# Load dataset from CSV file
file_path = '/content/AdaptiveMLaaS.csv'
client_data = pd.read_csv(file_path)

# Check required columns
if 'services' not in client_data.columns or 'Composability_Score' not in client_data.columns:
    raise ValueError("The dataset must contain 'services' and 'Composability_Score' columns.")

# Epsilon-Greedy Implementation
def epsilon_greedy(services, epsilon=0.1, iterations=200):
    """
    Epsilon-Greedy algorithm for selecting service triplets using composability scores.
    """
    # Initialize reward estimates and counts
    reward_estimates = {}
    pull_counts = {}

    # Generate all possible triplets of services
    all_triplets = list(combinations(services, 3))
    for triplet in all_triplets:
        reward_estimates[triplet] = 0
        pull_counts[triplet] = 0

    # Run Epsilon-Greedy for the specified number of iterations
    adjusted_iterations = len(services) * 20  # Scale iterations based on sample size
    for _ in range(adjusted_iterations):
        if random.random() < epsilon:
            # Exploration: Select a random triplet
            selected_triplet = random.choice(all_triplets)
        else:
            # Exploitation: Select the triplet with the highest estimated reward
            selected_triplet = max(reward_estimates, key=reward_estimates.get)

        # Evaluate the selected triplet using the mean composability score
        rewards = client_data[client_data['services'].isin(selected_triplet)]['Composability_Score'].values
        reward = rewards.mean()

        # Update reward estimates
        pull_counts[selected_triplet] += 1
        n = pull_counts[selected_triplet]
        reward_estimates[selected_triplet] += (reward - reward_estimates[selected_triplet]) / n

    # Sort triplets by estimated rewards
    sorted_triplets = sorted(reward_estimates.items(), key=lambda x: x[1], reverse=True)
    return sorted_triplets

# Define sample sizes
sample_sizes = [10, 20, 40, 60, 80, 100, 120, 150, 180]

# List to store results
computation_results = []

# Run the experiment for each sample size
for size in sample_sizes:
    subset_data = client_data.head(size)
    services = list(subset_data['services'])

    # Start measuring time
    start_time = time.time()

    # Run Epsilon-Greedy algorithm with iterations dynamically adjusted inside the function
    sorted_triplets = epsilon_greedy(services, epsilon=0.3)

    # End measuring time
    end_time = time.time()

    # Calculate computation time
    computation_time = end_time - start_time

    # Store the results
    computation_results.append((size, computation_time))

# Convert the results into a DataFrame
results_df_ep = pd.DataFrame(computation_results, columns=['Sample_Size', 'Computation_Time'])
results_df_ep
```

**Genetic Algorithm**
---

```python
import random
from itertools import combinations
import time
import pandas as pd

# Load dataset from CSV file
file_path = '/content/AdaptiveMLaaS.csv'
client_data = pd.read_csv(file_path)

# Initialize population with triplets of services
def initialize_population(services, population_size):
    return [random.sample(services, 3) for _ in range(population_size)]

# Fitness function using the mean composability score
def fitness_function(chromosome):
    scores = client_data[client_data['services'].isin(chromosome)]['Composability_Score'].values
    return scores.mean()

# Selection: Roulette wheel selection
def roulette_wheel_selection(population, fitness_scores):
    total_fitness = sum(fitness_scores)
    probabilities = [score / total_fitness for score in fitness_scores]
    return random.choices(population, weights=probabilities, k=2)

# Crossover: Swap one client between parents
def crossover(parent1, parent2):
    idx = random.randint(0, 2)
    offspring1 = parent1.copy()
    offspring2 = parent2.copy()
    offspring1[idx] = parent2[idx]
    offspring2[idx] = parent1[idx]
    return offspring1, offspring2

# Mutation: Replace one client in a triplet
def mutate(chromosome, services, mutation_rate=0.2):
    if random.random() < mutation_rate:
        idx = random.randint(0, 2)
        chromosome[idx] = random.choice(services)
    return chromosome

# GA implementation
def genetic_algorithm(services, population_size, generations, crossover_rate, mutation_rate):
    # Step 1: Initialize population
    population = initialize_population(services, population_size)

    for generation in range(generations):
        # Step 2: Evaluate fitness
        fitness_scores = [fitness_function(chromosome) for chromosome in population]

        # Step 3: Create new population
        new_population = []
        for _ in range(population_size // 2):
            # Selection
            parent1, parent2 = roulette_wheel_selection(population, fitness_scores)

            # Crossover
            if random.random() < crossover_rate:
                offspring1, offspring2 = crossover(parent1, parent2)
            else:
                offspring1, offspring2 = parent1, parent2

            # Mutation
            offspring1 = mutate(offspring1, services, mutation_rate)
            offspring2 = mutate(offspring2, services, mutation_rate)

            new_population.extend([offspring1, offspring2])

        # Replace old population
        population = new_population

    # Step 4: Evaluate final population and filter by reward > 0.7
    final_population_scores = [
        (chromosome, fitness_function(chromosome))
        for chromosome in population
        if fitness_function(chromosome) > 0.7
    ]

    # Step 5: Sort all combinations by fitness
    sorted_population = sorted(final_population_scores, key=lambda x: x[1], reverse=True)

    return sorted_population

# Define sample sizes and population sizes
sample_sizes = [10, 20, 40, 60, 80, 100, 120, 150, 180]
population_sizes = [1000,1000,1000,1000,1000,1000,1500,2000,2500,3000]

# List to store results
computation_results = []

# Run the experiment for each sample size with corresponding population size
for size, pop_size in zip(sample_sizes, population_sizes):
    subset_data = client_data.head(size)
    services = list(subset_data['services'])

    # Adjust the population size slightly with a 5% increase for diversity
    adjusted_population_size = int(pop_size * 1.05)
    generations = 100 + size  # Adjust generations dynamically based on sample size

    # Start measuring time
    start_time = time.time()

    # Run Genetic Algorithm
    sorted_population = genetic_algorithm(
        services, adjusted_population_size, generations, crossover_rate=0.8, mutation_rate=0.2
    )

    # End measuring time
    end_time = time.time()

    # Calculate computation time
    computation_time = end_time - start_time

    # Store the results
    computation_results.append((size, computation_time))

# Convert the results into a DataFrame
results_df_ga = pd.DataFrame(computation_results, columns=['Sample_Size', 'Computation_Time'])

results_df_ga
```
